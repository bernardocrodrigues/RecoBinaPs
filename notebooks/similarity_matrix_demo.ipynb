```python
import numpy as np
from surprise import Dataset
from lib.BooleanMatrixBasedRecomenders import get_similarity_matrix, jaccard_distance, cosine_distance
from tests.ToyDatasets import belohlavek_binary_dataset, zaki_binary_dataset, my_toy_binary_dataset

from rich.jupyter import print

import matplotlib.pyplot as plt
```

```python
belohlavek_similary_matrix = get_similarity_matrix(belohlavek_binary_dataset)
my_toy_similary_matrix = get_similarity_matrix(my_toy_binary_dataset)
zaki_similary_matrix = get_similarity_matrix(zaki_binary_dataset)

print(belohlavek_similary_matrix)
print(my_toy_similary_matrix)
print(zaki_similary_matrix)
```

```python
fig, axs = plt.subplots(1, 3, figsize=(20, 5))
fig.suptitle('Jaccard based Similartiy Matrices')

subfig = axs[0].imshow(belohlavek_similary_matrix, vmin=0, vmax=1, cmap='Greys')
axs[0].set(xlabel='User', ylabel='User', title='Belohlavek dataset')
fig.colorbar(subfig, ax=axs[0]).set_label('Similarity')

subfig = axs[1].imshow(my_toy_similary_matrix, vmin=0, vmax=1, cmap='Greys')
axs[1].set(xlabel='User', ylabel='User', title='My Toy dataset')
fig.colorbar(subfig, ax=axs[1]).set_label('Similarity')

subfig = axs[2].imshow(zaki_similary_matrix, vmin=0, vmax=1, cmap='Greys')
axs[2].set(xlabel='User', ylabel='User', title='Zaki dataset')
fig.colorbar(subfig, ax=axs[2]).set_label('Similarity')

fig.show()
```

```python
belohlavek_similary_matrix = get_similarity_matrix(belohlavek_binary_dataset, cosine_distance)
my_toy_similary_matrix = get_similarity_matrix(my_toy_binary_dataset, cosine_distance)
zaki_similary_matrix = get_similarity_matrix(zaki_binary_dataset, cosine_distance)
```
